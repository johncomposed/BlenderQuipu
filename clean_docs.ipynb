```python
def path_to_title(path):
    parts = path.split('/')[3:]
    formatted_parts = []
    for part in parts:
        # Remove file extension, replace underscores with spaces, capitalize
        part = part.split('.')[0].replace('_', ' ').title()
        formatted_parts.append(part)
    
    return ' > '.join(formatted_parts)

def path_to_slug(path):
    parts = path.split('/')[3:]
    formatted_parts = []
    for part in parts:
        # Remove file extension, replace underscores with spaces, capitalize
        part = part.split('.')[0]
        formatted_parts.append(part)
    
    return '-'.join(formatted_parts)


def file_to_elements(path):
    customTag = Title(
        text=path_to_title(path), 
        metadata=ElementMetadata.from_dict({"filename": path})
    )

    html_document = HTMLDocument.from_file(path)
    cleaned = html_document.doc_after_cleaners(skip_headers_and_footers=True)
    elements = html_document.pages[0].elements
    elements.insert(0, customTag)
    return elements

```

```python
for path in html_files:
    elements = file_to_elements(path)
    out = root_dir + '/element_json/' + path_to_slug(path) + '.json'
    print(path, len(elements), out)
    elements_to_json(elements, filename=out)
```
